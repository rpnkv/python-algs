```python
h_values_max=500
h_values=[val + 1 for val in range(0,h_values_max, 5)]
h_values_total=len(h_values)
```

```python
#seconds. starts at 3 usecs, end at 3 secs.
import math
start_value=3*math.pow(10, -3)
end_value=3
step=(end_value-start_value) / h_values_total
multiplier = int(math.pow(10, 5))
time_series=[int_val / multiplier for int_val in range(int(start_value*multiplier), int(end_value*multiplier), int(step*multiplier))]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ваши данные (пример)
#h_values = np.random.uniform(1, 500, 100)  # Количество вложенных списков
#time_series = np.random.uniform(0.003, 2.98, 100)  # Время выполнения

# Создаем график с нужным размером
plt.subplots(figsize=(15, 4))

# Устанавливаем пределы осей с отступами
x_min, x_max = min(h_values), max(h_values)
y_min, y_max = min(time_series), max(time_series)

# Добавляем отступы (10% от диапазона)
x_padding = (x_max - x_min) * 0.1
y_padding = (y_max - y_min) * 0.1

plt.xlim(x_min - x_padding, x_max + x_padding)
plt.ylim(y_min - y_padding, y_max + y_padding)

# Явно указываем минимальные и максимальные значения на осях
plt.xticks([x_min, *range(10, 450, 50), x_max])
plt.yticks([y_min, 0.5, 1.0, 1.5, 2.0, 2.5, y_max])

# Рисуем график
plt.plot(h_values, time_series, 'r-', )#linewidth=1, markersize=4)
plt.xlabel('Количество вложенных списков')
plt.ylabel('Время выполнения, сек')

# Добавляем точки для минимальных и максимальных значений
plt.scatter([x_min, x_max], [y_min, y_max], color='blue', s=50, zorder=5)
plt.scatter(x_min, y_min, color='blue', s=50, zorder=5, label=f'Min: ({x_min}, {y_min:.3f})')
plt.scatter(x_max, y_max, color='red', s=50, zorder=5, label=f'Max: ({x_max}, {y_max:.3f})')

# Добавляем легенду
plt.legend()

# Включаем сетку для лучшей читаемости
plt.grid(True, alpha=0.3)

plt.title("Зависимость времени выполнения от количества вложенных списков")
#plt.tight_layout()
plt.show()
```
